```python
# Universal Islamic AI Demo
!pip install -q transformers torch pandas numpy

from transformers import pipeline

# Load universal model
classifier = pipeline("text-classification", model="ADANiD/islamic-ai-foundation")

# Test examples
examples = [
    "بِسْمِ اللَّهِ الرَّحْمَٰنِ الرَّحِيمِ",
    "حدثنا عبد الله بن يوسف قال أخبرنا مالك...",
    "کیا وضو کے بغیر نماز پڑھی جا سکتی ہے؟"
]

for text in examples:
    result = classifier(text)
    print(f"Text: {text[:30]}...")
    print(f"Classification: {result[0]['label']} ({result[0]['score']:.2f})\n")
```
